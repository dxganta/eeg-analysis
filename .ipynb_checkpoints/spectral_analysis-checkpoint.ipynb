```python
import numpy as np
from numpy.fft import fft, ifft
%matplotlib inline
import matplotlib.pyplot as plt
from autoreject import get_rejection_threshold
import mne
```

```python
eeg = np.genfromtxt("first-recording-resting.csv", delimiter=",", skip_header=1, usecols=(0,1,2,3,4))
channel_labels = np.genfromtxt("first-recording-resting.csv", delimiter=",", usecols=(1,2,3,4), max_rows=1, dtype='str')
srate = 256 # sampling rate of the muse 2 headset

channel_labels
```

```python
def label_to_ind(label):
    '''
    takes the string label of a channel
    and returns its index position
    '''
    return np.where(np.char.lower(channel_labels) == label.lower())[0][0]

label_to_ind('af8')
```

```python
times = eeg[:, 0]
times  = times - np.min(times) # start time stamps from zero
data = eeg[:, 1:]
npnts = len(times)
data.shape
```

```python
times
```

```python
plt.figure(figsize=(16,8))
plt.plot(times, data, label=channel_labels)
plt.xlabel("Time (in seconds)")
plt.ylabel("Voltage(uV)")
plt.legend()
plt.show()
```

```python
ch_to_ind = label_to_ind('af7')
hz = np.linspace(0, srate/2, npnts//2+1)

dataX = fft(data, axis=0)
pwr = (2*abs(dataX / npnts)) ** 2
plt.figure(figsize=(16,8))
plt.plot(hz, pwr[:len(hz), :], label=channel_labels)
plt.title(f'Amplitude from all channels')
plt.xlabel('Frequency(Hz))')
plt.ylabel('Power')
plt.xlim([0,5])
plt.ylim([0,300])
plt.legend()
plt.show()
```

# Using MNE

```python
# transform data into mne raw format
info = mne.create_info(ch_names=list(channel_labels), sfreq=srate, ch_types=['eeg']*len(channel_labels))
raw = mne.io.RawArray(data.T, info)

# plotting the data
plt.figure(figsize=(16,8))
plt.plot(raw.get_data(picks='AF7').T, 'k')
plt.show()

raw.set_montage("standard_1005")

raw.plot_sensors(show_names=True);

raw.compute_psd(fmax=100).plot();
```

# Band-pass Filtering data 

```python
# Band pass filter data to be between 0.1 and 40 Hz

low_cut = 0.1
high_cut = 40
raw_filt = raw.copy().filter(low_cut, high_cut);
raw_filt.compute_psd(fmax=60).plot();
```

```python
# plotting the data
plt.figure(figsize=(16,8))
plt.plot(raw.get_data(picks='TP10').T, 'k')
plt.plot(raw_filt.get_data(picks='TP10').T)
plt.xlim([0,500])
plt.show()
```

# Preprocessing

```python
from sklearn.decomposition import FastICA

# Define FastICA transformer
transformer = FastICA(n_components=4, random_state=0)

# Apply FastICA to the data (excluding timestamps and AUX channel)
ica_sources = transformer.fit_transform(data)

# Plot the independent components
plt.figure(figsize=(15, 10))
for i in range(4):
    plt.plot(times, ica_sources[ :, i], label=channel_labels[i])

plt.title('Independent Components of EEG Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

# Time Frequency Analysis Manually

```python
def time_frequency(data_, times, freqrange=[1,60], numfrex=43, srate=256, channel_labels=None):
    '''
        creates a time frequency power plot of the data_ and plots it for
        every channel
        
        data -> of shape channels x time
        times -> the times array for the given data_
        freqrange -> extract only these frequencies (in Hz)
        numfrex -> number of frequencies between lowest and highest
    '''
    pi = np.pi
    
    assert data_.shape[0] < data_.shape[1], "data shape incorrect"
    assert channel_labels is None or len(channel_labels) == data_.shape[0], "channel_labels must be of same length as number of channels"

    # set up convolution parameters
    wavtime = np.arange(-2,2-1/srate,1/srate)
    frex    = np.linspace(freqrange[0],freqrange[1],numfrex)
    nData   = data_.shape[1];
    nKern   = len(wavtime);
    nConv   = nData + nKern - 1;
    halfwav = (len(wavtime)-1)//2;

    # number of cycles
    numcyc = np.linspace(3,15,numfrex);

    # create wavelets
    cmwX = np.zeros((numfrex, nConv), dtype=complex)
    for fi in range(numfrex):

        # create time-domain wavelet
        s = numcyc[fi] / (2*pi*frex[fi])
        twoSsquared = (2*s) ** 2
        cmw = np.exp(2*1j*pi*frex[fi]*wavtime) * np.exp( (-wavtime**2) / twoSsquared )


        # compute fourier coefficients of wavelet and normalize
        cmwX[fi, :] = fft(cmw, nConv)
        cmwX[fi, :] = cmwX[fi, :] / max(cmwX[fi, :])

    # initialize time-frequency output matrix
    tf = np.zeros((data_.shape[0], numfrex, data_.shape[1])) # channels X frequency X times

    # loop over channels
    for chani in range(data_.shape[0]):

        # compute Fourier coefficients of EEG data (doesn't change over frequency!)
        eegX = fft(data_[chani, :] , nConv)

        # perform convolution and extract power (vectorized across frequencies)
        as_ = ifft(cmwX * eegX[None, :], axis=1)
        as_ = as_[:, halfwav: -halfwav]
        tf[chani, :, :] = np.abs(as_) ** 2
        
    tf = np.mean(tf, axis=0)

    plt.figure(figsize=(15,6))
    plt.contourf(times,frex, tf, 40, cmap='jet') # changed colormap to 'viridis'
    plt.xlabel('Time (s)')
    plt.ylabel('Frequencies (Hz)')
    plt.title( f'Average power across all channels')
    plt.show()

time_frequency(data.copy().T, times)
```
